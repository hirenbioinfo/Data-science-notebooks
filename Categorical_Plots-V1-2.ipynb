A categorical variable (also known as a qualitative variable) in statistics is a variable that has a finite number of potential values, allocating each human or other unit of observation to a specific group or nominal category based on some qualitative attribute. Categorical variables are known as enumerations or enumerated types in computer science and certain disciplines of mathematics. Each of a categorical variable's potential values is sometimes referred to as a level (but not in this article). A categorical distribution is the probability distribution associated with a random categorical variable.

Differnt types of categorical plots are bellow:

* boxplot
* violinplot
* stripplot
* swarmplot
* barplot
* countplot


```python
import seaborn as sns
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
tips = pd.read_csv('tips.csv')
tips.head()
#tips.describe
```

## barplot and countplot

These very similar plots allow you to get aggregate data off a categorical feature in your data. **barplot** is a general plot that allows you to aggregate the categorical data based off some function, by default the mean:

```python
sns.set_theme(style="whitegrid")
sns.barplot(x='sex',y='total_bill',data=tips,palette = 'hls',capsize = 0.05,             
            saturation = 8)
```

You can change the estimator object to your own function, that converts a vector to a scalar:

```python
sns.set_theme(style="whitegrid")
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std,palette = 'hls',capsize = 0.05,             
            saturation = 8)
```

### countplot
<https://seaborn.pydata.org/generated/seaborn.countplot.html>

A count plot is a histogram for a category variable. So you can compare counts across nested variables using the same API and settings as barplot().This is essentially the same as barplot except the estimator is explicitly counting the number of occurrences. Which is why we only pass the x value:

```python
sns.set_theme(style="whitegrid")
sns.countplot(x='sex',data=tips,palette = 'hls',capsize = 0.05,             
            saturation = 8)
```

## boxplot and violinplot
<https://seaborn.pydata.org/generated/seaborn.boxplot.html>
<https://seaborn.pydata.org/generated/seaborn.violinplot.html>

boxes and violinplots used to depict the distribution of categorical data in a visual manner. Data may be compared across different variables or levels of a categorical variable using a box plot (or box-and-whisker plot). To illustrate the whole distribution, except for the points that are considered to be "outliers" using a technique that takes into account the interquartile range, a box is used.

```python
sns.set_theme(style="whitegrid")
sns.boxplot(x="day", y="total_bill", data=tips,palette='rainbow',width=0.5)
# if you wish you can change the color paletee 
#colors = ['#78C850', '#F08030', '#6890F0','#F8D030']
```

```python
# Can do entire dataframe with orient='h'
sns.set_theme(style="whitegrid")
sns.boxplot(data=tips,palette='rainbow',orient='h')
```

### Grouped boxplots
A grouped boxplot is a boxplot where categories are organized in groups and subgroups.

```python
sns.set_theme(style="whitegrid")
sns.boxplot(x="day", y="total_bill", hue="smoker",data=tips, palette="pastel")
```

### Violinplot
<https://seaborn.pydata.org/generated/seaborn.violinplot.html>

A violin plot is similar to a box and whisker plot in that it serves the same purpose. One (or more) categorical variables are used to demonstrate how quantitative data is distributed across many levels of a category variable. Instead of using a box plot, a violin plot uses a kernel density estimate of the underlying distribution to provide a visual representation.

```python
sns.set_theme(style="whitegrid")
sns.violinplot(x="day", y="total_bill", data=tips,palette='rainbow')
```

```python
sns.set_theme(style="whitegrid")
sns.violinplot(x="day", y="total_bill", data=tips,hue='sex',palette="muted")
```

```python
sns.violinplot(x="day", y="total_bill", data=tips,hue='sex',split=True,palette="muted")
```

## stripplot and swarmplot
A scatterplot with one categorical variable will be drawn using the stripplot. In circumstances when you wish to display all observations together with some representation of the underlying distribution, a strip plot may be used alone or as an addition to a box or violin plot.

Similar to the strip plot, except the points are shifted so that they don't overlap with each other on just one axis (the category axis). This is a more accurate depiction of the distribution of values, but it does not scale well to large numbers of observations (both in terms of the ability to show all the points and in terms of the computation needed to arrange them).

```python
sns.set_theme(style="whitegrid")
sns.stripplot(x="day", y="total_bill", data=tips)
```

```python
sns.set_theme(style="whitegrid")
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True)
```

```python
sns.set_theme(style="whitegrid")
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True,hue='sex',palette='Set1')
```

```python
sns.set_theme(style="whitegrid")
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True,hue='sex',palette='Set1',split=True)
```

```python
sns.set_theme(style="whitegrid")
sns.swarmplot(x="day", y="total_bill", data=tips)
```

```python
sns.set_theme(style="whitegrid")
sns.swarmplot(x="day", y="total_bill",hue='sex',data=tips, palette="Set1", split=True)
```

### Combining Categorical Plots

```python
sns.set_theme(style="whitegrid")
sns.violinplot(x="tip", y="day", data=tips,palette='rainbow')
sns.swarmplot(x="tip", y="day", data=tips,color='black',size=3)
```
